```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

1. Scatter plot: Seaborn's scatter plot is used to plot the relationship between two variables. It is used to visualize the correlation between two variables and to identify outliers.

2. Line plot: Seaborn's line plot is used to plot the trend of a variable over time. It is used to visualize the change in a variable over time.

3. Histogram: Seaborn's histogram plot is used to plot the distribution of a variable. It is used to visualize the distribution of data, whether it is skewed or symmetrical.

4. Bar plot: Seaborn's bar plot is used to compare the data between different groups. It is used to visualize the difference between groups.

5. Heat map: Seaborn's heat map plot is used to plot the correlation between multiple variables. It is used to visualize the correlation between multiple variables at once.

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python

# Load the fmri dataset
fmri_data = sns.load_dataset("fmri")
```

```python

# Plot line plot for different events and regions
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.Note: pclass, age, and fare are columns in the titanic dataset.

```python
# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")
```

```python

# Plot boxplot for age based on pclass
sns.boxplot(x="pclass", y="age", data=titanic_data)

# Plot boxplot for fare based on pclass
sns.boxplot(x="pclass", y="fare", data=titanic_data)
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
# Load the diamonds dataset
diamonds_data = sns.load_dataset("diamonds")
```

```python

# Plot histogram of price based on cut
sns.histplot(data=diamonds_data, x="price", hue="cut", multiple="stack")
```

### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
# Load the iris dataset
iris_data = sns.load_dataset("iris")
```

```python
# Plot pair plot with hue parameter for species column
sns.pairplot(data=iris_data, hue="species")
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python

# Load the flights dataset
flights_data = sns.load_dataset("flights")
```

```python


# Reshape the data into a pivot table
flights_pivot = flights_data.pivot("month", "year", "passengers")

# Plot a heatmap
sns.heatmap(flights_pivot, cmap="YlGnBu")
```
