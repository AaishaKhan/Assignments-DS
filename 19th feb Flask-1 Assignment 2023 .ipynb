#### Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a Python micro web framework that allows developers to create flexible and lightweight online applications. Flask is regarded as a minimalist framework since it offers only the tools and libraries required to construct a web application, allowing developers free to select and integrate other libraries as needed.

* Flask is meant to be lightweight and flexible, allowing developers to construct web applications that meet their individual needs and constraints.


* Flask features a straightforward and easy-to-understand syntax, making it an excellent alternative for developers new to web development or the Python programming language.

* Built-in Development Server: Flask includes a development server that helps developers to rapidly and easily test their apps.

    
* Flask has a strong developer community and a diverse ecosystem of plugins and extensions, making it simple to discover answers to common web development difficulties.


* Scalable: Flask is built to be scalable, allowing developers to create web applications that can manage a high volume of users and requests.

-: Overall, Flask is an excellent solution for rapidly and simply developing small to medium-sized web applications while retaining flexibility and customisation.

#### Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello, World!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

#### Q3. What is App routing in Flask? Why do we use app routes?

The technique of mapping a URL to a specific function in your application is referred to as routing in Flask. The act of establishing these URL mappings within your Flask application is known as app routing.

* When a user visits a given URL and makes a request to your Flask application, Flask utilises the established routes to identify which function should handle the request. The function is then run with any required arguments, such as URL parameters or request data, and returns a response to the user's browser.

* App routes are used to define the behavior of your Flask application when a specific URL is requested. For example, if a user visits the URL /home, you may want to display a page with information about your application. 

* App routes allow you to define the behavior of your Flask application for different URLs, and enable you to create dynamic applications that can respond to user input.

#### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

The url_for() function in Flask is used for URL building. It generates a URL to a specific function based on the endpoint name. It takes the endpoint name as its first argument and any number of keyword arguments, each corresponding to a variable part of the URL rule

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Welcome to the home page!'

@app.route('/about')
def about():
    return 'This is the about page.'

@app.route('/profile/<username>')
def profile(username):
    return f'This is {username}\'s profile.'

with app.test_request_context():
    print(url_for('home'))
    print(url_for('about'))
    print(url_for('profile', username='Ash K'))
    
if __name__ == '__main__':
 app.run(host="0.0.0.0")    
```
